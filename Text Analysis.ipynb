```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import scipy.stats as stats
from collections import Counter
from gensim import corpora
from gensim.models import LdaModel
import spacy
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics.pairwise import cosine_similarity
import openpyxl

# Ensure the necessary NLTK corpora are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Load data from Excel
data = pd.read_excel('Surveyquestion.xlsx', sheet_name='Questions')

# Remove incomplete or irrelevant responses
data.dropna(subset=['Questions_text'], inplace=True)

# Standardize data for consistency
data['Questions_text'] = data['Questions_text'].str.lower().str.strip()

# Thematic Analysis Function
def thematic_analysis(question):
    if "price" in question:
        return "Price"
    elif "feature" in question:
        return "Features"
    elif "review" in question:
        return "Reviews"
    elif "brand" in question:
        return "Brands"
    elif "warranty" in question:
        return "Warranty"
    elif "discount" in question:
        return "Discounts"
    else:
        return "General"

data['Thematic_Theme'] = data['Questions_text'].apply(thematic_analysis)

# Descriptive Statistics
# Calculate frequency and percentage of each question
question_counts = data['Questions_text'].value_counts()
question_percentages = question_counts / len(data) * 100

# Identify the top 10 most frequently asked questions
top_10_questions = question_counts.head(10)
print("Top 10 Questions:\n", top_10_questions)

# Thematic Analysis
stop_words = set(stopwords.words('english'))
data['tokens'] = data['Questions_text'].apply(lambda x: [word for word in word_tokenize(x) if word.isalpha() and word not in stop_words])

# Cluster Analysis
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['Questions_text'])
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
data['Cluster'] = kmeans.labels_

# Visualize clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x=X.toarray()[:, 0], y=X.toarray()[:, 1], hue=data['Cluster'])
plt.title('Question Clusters')
plt.savefig('question_clusters.png')
plt.show()

# Sentiment Analysis
analyzer = SentimentIntensityAnalyzer()
data['Sentiment'] = data['Questions_text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
data['Sentiment_Category'] = pd.cut(data['Sentiment'], bins=[-1, -0.05, 0.05, 1], labels=['negative', 'neutral', 'positive'])
print("Sentiment Categories:\n", data['Sentiment_Category'].value_counts())

# Visualize Sentiment Categories
plt.figure(figsize=(10, 7))
sns.countplot(x='Sentiment_Category', data=data, palette='coolwarm')
plt.title('Sentiment Categories')
plt.savefig('sentiment_categories.png')
plt.show()

# Comparative Analysis
cross_tab = pd.crosstab(data['Cluster'], data['Sentiment_Category'])
chi2, p, dof, ex = stats.chi2_contingency(cross_tab)
print("Chi-square test result:", chi2, "p-value:", p)

# Content Analysis
all_words = [word for tokens in data['tokens'] for word in tokens]
word_counts = Counter(all_words)
common_words = word_counts.most_common(10)
print("Common Words:\n", common_words)

# Visualize Common Words
common_words_df = pd.DataFrame(common_words, columns=['Word', 'Frequency'])
plt.figure(figsize=(10, 7))
sns.barplot(x='Frequency', y='Word', data=common_words_df, palette='viridis')
plt.title('Most Common Words')
plt.savefig('common_words.png')
plt.show()

# Latent Dirichlet Allocation (LDA)
dictionary = corpora.Dictionary(data['tokens'])
corpus = [dictionary.doc2bow(tokens) for tokens in data['tokens']]
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)
topics = lda_model.print_topics(num_words=5)
print("LDA Topics:")
for topic in topics:
    print(topic)

# Add LDA topics to the DataFrame
lda_topics = []
for i, row in data.iterrows():
    bow = dictionary.doc2bow(row['tokens'])
    topic_distribution = lda_model.get_document_topics(bow)
    top_topic = max(topic_distribution, key=lambda x: x[1])[0]
    lda_topics.append(top_topic)

data['LDA_Topic'] = lda_topics

# Named Entity Recognition (NER)
nlp = spacy.load('en_core_web_sm')
data['Entities'] = data['Questions_text'].apply(lambda x: [(ent.text, ent.label_) for ent in nlp(x).ents])
print("Named Entities:\n", data['Entities'].head())

# Dependency Parsing
data['Dependency_Parse'] = data['Questions_text'].apply(lambda x: [(token.text, token.dep_) for token in nlp(x)])
print("Dependency Parsing:\n", data['Dependency_Parse'].head())

# Text Classification
data['Category_ID'] = data['Cluster'].factorize()[0]
X_train, X_test, y_train, y_test = train_test_split(X, data['Category_ID'], test_size=0.2, random_state=0)
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print("Text Classification Report:\n", classification_report(y_test, y_pred))

# Semantic Similarity Analysis using TF-IDF
vectorizer = TfidfVectorizer().fit(data['Questions_text'])
tfidf_matrix = vectorizer.transform(data['Questions_text'])
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("Semantic Similarity Matrix:\n", similarity_matrix)

# Save results to Excel
with pd.ExcelWriter('questions_analysis_v1.xlsx') as writer:
    data.to_excel(writer, index=False, sheet_name='Analysis')
    common_words_df.to_excel(writer, index=False, sheet_name='Common_Words')
    pd.DataFrame(similarity_matrix).to_excel(writer, index=False, sheet_name='Similarity_Matrix')

print("Analysis saved to questions_analysis_v1.xlsx")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

# Assuming 'data' is your DataFrame and 'Cluster' is your class column
def get_top_words_per_class(data, class_column, text_column, num_words=10):
    vectorizer = TfidfVectorizer(stop_words='english')
    data['tfidf'] = list(vectorizer.fit_transform(data[text_column]).toarray())
    feature_names = vectorizer.get_feature_names_out()
    
    top_words = {}
    for cluster in data[class_column].unique():
        cluster_data = data[data[class_column] == cluster]
        tfidf_matrix = np.array(cluster_data['tfidf'].tolist())
        average_tfidf = np.mean(tfidf_matrix, axis=0)
        top_indices = average_tfidf.argsort()[-num_words:][::-1]
        top_features = [feature_names[i] for i in top_indices]
        top_words[cluster] = top_features
    
    return top_words

top_words_per_class = get_top_words_per_class(data, 'Cluster', 'Questions_text')
print(top_words_per_class)
```

```python
def print_example_texts(data, class_column, text_column, num_examples=5):
    for cluster in data[class_column].unique():
        print(f"\nClass {cluster} Examples:")
        examples = data[data[class_column] == cluster][text_column].head(num_examples).tolist()
        for example in examples:
            print(f"- {example}")

print_example_texts(data, 'Cluster', 'Questions_text')
```
