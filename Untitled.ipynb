```python
import os
from PIL import Image

# Paths
input_dir = 'images'
output_dir = 'journal_images'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Supported image extensions
types = ('.png', '.jpg', '.jpeg', '.tif', '.tiff', '.bmp')

for filename in os.listdir(input_dir):
    if filename.lower().endswith(types):
        img_path = os.path.join(input_dir, filename)
        
        # Open the image
        with Image.open(img_path) as img:
            # Convert to RGB if necessary (PDFs don't support RGBA)
            if img.mode in ('RGBA', 'P'):
                img = img.convert('RGB')

            # 1) Update original image DPI metadata to 1200
            img.save(img_path, dpi=(1200, 1200))

            # 2) Save as PDF with 1200 DPI
            pdf_name = os.path.splitext(filename)[0] + '.pdf'
            pdf_path = os.path.join(output_dir, pdf_name)
            img.save(pdf_path, 'PDF', dpi=(1200, 1200))
            print(f'Saved PDF: {pdf_path}')

print('All images updated to 1200 DPI and saved as PDFs!')
```
